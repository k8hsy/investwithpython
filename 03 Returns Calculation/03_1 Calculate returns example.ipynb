3.1 Return Calculation Practice
   - Simple return
     - 수식 : (End Price/Beginning Price) -1
       파이썬의 데이타프레임 시계열 데이터에서는 .pct_change() 수식을 쓸 수 있음

   - Log return
     - 수식 : log(last price / first price) or
              log(last price)-log(first price)


```python
# Loading Python Packages to use 
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

#download Apple stock price from Yahoo finance and create dataframe
df=pd.DataFrame(yf.download("AAPL", start='2018-12-31')['Adj Close'])
df
```

```python
# Calculate simple and log returns from above dataframe

# Simple return Calculation(same results with different ways)
df['simple return1'] = df['Adj Close']/df['Adj Close'].shift()-1
df['simple return2'] = df['Adj Close'].pct_change(1)

# Log Return Calculation(same results with different ways)

df['logReturn1']=np.log(df['Adj Close']/df['Adj Close'].shift(1))
df['logReturn2']=np.log(df['Adj Close'])-np.log(df['Adj Close'].shift(1))
df['logReturn3']=np.log(1+df['Adj Close'].pct_change(1))


# Log return cumulative return
df['logRcsum']=df['logReturn1'].cumsum()
df['recalculate price']=np.exp(df['logRcsum'])*df['Adj Close'][0]
df
```

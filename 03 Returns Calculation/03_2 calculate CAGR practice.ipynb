3.2 Calculate CAGR Practice
   

```python
#Packages used (사용된 패키지)
import pandas as pd
import numpy as np
#import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CAGR calculation function. CAGR=(last value / first value)^(1/# of years) - 1 
def getCAGR(first_price, last_price, years):
    CAGR=((last_price/first_price)**(1/years)-1)
    return CAGR
first_price = 10000
last_price = 20000
No_of_years = 10

cagr=getCAGR(first_price=first_price, last_price=last_price, years=No_of_years)
print('CAGR = ',cagr)
```

```python
# 투자원금이 두배가 되는데 걸리는 시간에 대한 CAGR 구하기
c=[]
y1=[]
for i in range(1,21):
    cagr=getCAGR(first_price=first_price, last_price=last_price, years=i)
    c.append([cagr,i])
    y1.append(i)
df_CAGR_year=pd.DataFrame(c)
df_CAGR_year
```

```python
df_CAGR_year.columns=["CAGR","Total Years"]
df_CAGR_year
```

```python
x_value=df_CAGR_year['Total Years']
y_value=df_CAGR_year["CAGR"]

fig = plt.figure(figsize=(16,10)) ## 캔버스 생성
plt.plot(x_value,y_value)
plt.ylabel("CAGR")
plt.xlabel('Years')
plt.title('No of years to make double')
plt.show()
```

```python
# 10년 기간의 CAGR에 따르는 투자원금의 변화 구하기
c=[]
for i in range(15000,41000,1000):
    cagr=getCAGR(first_price=first_price, last_price=i, years=10)
    c.append([cagr,i,No_of_years])
df_CAGR_year10=pd.DataFrame(c)
df_CAGR_year10.columns=["CAGR","Total Amounts","years"]
df_CAGR_year10
```

```python

x_value=df_CAGR_year10['CAGR']
y_value=df_CAGR_year10['Total Amounts']

fig = plt.figure(figsize=(16,10)) 
plt.plot(x_value,y_value)
plt.xlabel("CAGR")
plt.ylabel('Final Amt')
plt.title('10Year CAGR returns')
plt.show()
```

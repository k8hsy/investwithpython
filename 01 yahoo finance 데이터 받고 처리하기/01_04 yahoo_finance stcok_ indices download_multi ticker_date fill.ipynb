1. The process of downloading major stock indices data provided by Yahoo Finance and processing as Dataframe
    - Continue from "01_03 yahoo_finance stock_indices download multi ticker.ipynb"
    - Create DataFrame with fill value    
    - Download multi historical price data from yahoo finance
        - Input : Historical price DataFrame
        - Output: Index DataFrames with fill value, Export Dataframe to excel file or csv file

```python
# Loading Python Packages to use 
import yfinance as yf
import pandas as pd
import numpy as np
import datetime as dt                

# use the code from "01_03 yahoo_finance stock_indices download multi ticker.ipynb"

major_indices = pd.DataFrame(pd.read_html("https://finance.yahoo.com/world-indices")[0])

# Slicing columns
i = 5                                       # set column counter variable, change i to slicing columns
major_indices1 = major_indices.iloc[:, :i]  # create another dataframe 'major_indices1'
major_indices1
drop_row_index_list = [3,4,5,7,11,12,13,14,20,25,28,31,33,34,35]  

major_indices2 = major_indices.drop(drop_row_index_list,axis=0)       
major_indices1_2 = major_indices1.drop(drop_row_index_list,axis=0) #Datafrmae to use 

Tks=major_indices1_2['Symbol'].values.tolist()

# Declare historical price variables[symbols('ticker'), date, interval, etc]
# yf.download(ticker, start, end, interval, etc) input varibales:'ticker', 'start=', 'end=', 'interval=' etc
tk = Tks                    # Use ticker list 'Tks' of 'Symbol' from DataFrame 'major_indices1_2' 

s_d = '2007-12-20'          # start date
e_d = '2022-12-31'          # end date, if it is up to today, just drop 'end' parameter
periods = 'ytd'             # Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
interval='1d'               # Valid interval : Daily;'1D', Weekly:'1wk',monthly:'1mo', 3month :'3mo'                        
# Download from start date to end date
df = pd.DataFrame(yf.download(tk,start=s_d,end=e_d,interval=interval)['Adj Close'])
df
```

2. Fill missing values
    - Due to the different financial market closed date, the downloaded data has 'NaN(Not a number)' values
    - To calculate and compare index returns, need to fill NaN value with privious value

```python
# 2.1 count Nan values
df.isnull().sum()
```

```python
#2.2 Reset date index to fill process

df1=df.reset_index()            # date필드 설정을 위해서 인덱스를 해제
df1
```

```python
# 2.3 Get all dates within the sample period (데이터에 없는 날짜 채우기)
start_date = df1['Date'].astype(str).iloc[0] # first date of Dataframe df1
end_date = df1['Date'].astype(str).iloc[-1]  # last date of Dataframe df1   
print(start_date)
print(end_date)

# date_range function include both sarting date and ending date, but yahoo download data from 1day before starting date to 1day before ending date

date_all = pd.date_range(start_date, end_date, freq='D').to_frame().rename(columns={0:'Date'}).reset_index(drop=True) # 기간 중 빈날짜 채우기
date_all
```

```python
#2.4 Merge two dataframes and fill NaN value with previous value
df1_all = pd.merge(date_all,df1,how='left')     # 두 개의 데이타프레임을 합쳐서 모든 날짜를 포함하고 빈 데이타는 다음 데이타로 채움
df1_all = df1_all.fillna(method='ffill')
df1_all
```

3. Create Year-end, month-end price Dataframes
    - Add last date in order to calculate YTD(Year to date), MTD(Month to date)

```python
#3.1 Create year-end price dataframe including last date and chage colume name from ticker to index name
year_end = pd.date_range(start_date, end_date, freq='Y')  # Year end 날짜 리스트를 구하기
df_y=pd.DataFrame(year_end)                     # 연말날자 리스트를 데이타프레임으로 만들기

df_y.loc[len(df_y)]=df1.iloc[-1,0]           #최종일자 행추가
df_y.columns=['Date']                        #column name set as 'date'

df_y=pd.merge(df_y,df1_all, how='left')    # 최종날자를 포함하고 있는 연말 날자 데이타프레임으로 합치기
df_y
```

```python
#3.2 Set year as index and change column name

df_y['Year']=df_y['Date'].dt.year           # 날자에서 연도항목을 추출하여 인덱스로 설정
df_y=df_y.set_index('Year', drop=True)      # 연도컬럼을 다시 인덱스로 설정
df_y=df_y.iloc[:,1:]          
     
col_list=df_y.columns.to_list()        # df_y 의 컬럼을 리스트로 만든다

df_name=pd.DataFrame(col_list)
df_name.columns=['Symbol']
df_name1=pd.merge(df_name, major_indices, how='left')
df_name1=df_name1['Name'].to_list()
df_y.columns=df_name1
df_y
```

```python
#3.3 #3.1 Create month-end price dataframe including last date and chage colume name from ticker to index name
month_end = pd.date_range(s_d, e_d, freq='M')  # 월말 날짜 리스트를 구하기
df_m=pd.DataFrame(month_end)                    # 월말날자 리스트를 데이타프레임으로 만들기                                         
df_m.loc[len(df_m)]=df1.iloc[-1,0]   
df_m.columns=['Date']                
df_m=pd.merge(df_m,df1_all, how='left')    # 최종날자를 포함하고 있는 월말 날자 데이타프레임으로 합치기

df_m['Year']=df_m['Date'].dt.year
df_m['Month']=df_m['Date'].dt.month          # 날자에서 월항목을 추출하여 인덱스로 설정

df_m
```

```python
#3.4 Set month as index and change column name

df_mm=df_m.set_index(['Year','Month'], drop=True)      # Year, month columns set as index
df_mm=df_mm.iloc[:,1:]                                 # exclude date column
     
col_list_m=df_mm.columns.to_list()        # df_m 의 컬럼을 리스트로 만든다

df_name_m=pd.DataFrame(col_list_m)
df_name_m.columns=['Symbol']
df_name1_m=pd.merge(df_name_m, major_indices, how='left')
df_name1_m=df_name1_m['Name'].to_list()
df_mm.columns=df_name1_m
df_mm
```

4. Export Dataframe to CSV file or Excel file
    - df1_all, df_y, df_m, df_mm 

```python
# 4.1 CSV file and Excel file
df1_all.to_csv('01_04 Index historical daily price for all date.csv')
df_y.to_csv('01_04 Year-end index price.csv')
df_m.to_csv('01_04 Month-end index price.csv')
df_mm.to_csv('01_04 Month-end by year index price.csv')

df1_all.to_excel('01_04 Index historical daily price for all date.xlsx')
df_y.to_excel('01_04 Year-end index price.xlsx')
df_m.to_excel('01_04 Month-end index price.xlsx')
df_mm.to_excel('01_04 Month-end by year index price.xlsx')
```

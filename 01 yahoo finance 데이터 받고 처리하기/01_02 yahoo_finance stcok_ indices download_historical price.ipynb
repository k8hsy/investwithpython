1. The process of downloading major stock indices data provided by Yahoo Finance and processing as Dataframe
    - Continue from "01_01 yahoo_finance stock_indices download.ipynb"
        - import additional libraries
    - Download historical price data from yahoo finance
        - Input : index ticker, periods
        - Output: DataFrames of historical data (Open,High,Low,Close,Adj Close,Volume)
    - Draw simple line graph

```python
# Loading Python Packages to use 
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt                     # Library for graphs
import seaborn as sns                               # Library for graphs
```

```python
# use the code from '"01_01 yahoo_finance stock_indices download.ipynb"

major_indices = pd.DataFrame(pd.read_html("https://finance.yahoo.com/world-indices")[0])

# Slicing columns
i = 5                                       # set column counter variable, change i to slicing columns
major_indices1 = major_indices.iloc[:, :i]  # create another dataframe 'major_indices1'
major_indices1
drop_row_index_list = [4,5,7,11,12,13,14,20,25,28,31,33,34,35]  

major_indices2=major_indices.drop(drop_row_index_list,axis=0)       
major_indices1_2=major_indices1.drop(drop_row_index_list,axis=0)
major_indices1_2
```

2. Input variables

```python
# Declare historical price variables[symbols('ticker'), date, interval, etc]
# yf.download(ticker, start, end, interval, etc) input varibales:'ticker', 'start=', 'end=', 'interval=' etc
tk = '^KS11'                # Refer 'Symbol' from DataFrame 'major_indices1_2' for ticker
s_d = '2019-12-20'          # start date
e_d = '2022-12-31'          # end date, if it is up to today, just drop 'end' parameter
periods = 'ytd'             # Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
interval='1d'               # Valid interval : Daily;'1D', Weekly:'1wk',monthly:'1mo', 3month :'3mo'                        
```

3. Histrical price download [change variables :'tk', 's_d', 'e_d', 'periods', 'interval'] 

```python
# 3.1_1 YTD 
df_ytd=pd.DataFrame(yf.download(tk,period=periods,interval=interval))
# df_ytd=pd.DataFrame(yf.download(tk,period='ytd')) # instead of use variable, directly input period = 'ytd'
df_ytd
```

```python
# 3.1_2 Download only 'Adj Close' column
df_ytd_adj_c=pd.DataFrame(yf.download(tk,period='ytd',interval=interval)['Adj Close'])
df_ytd_adj_c
```

```python
# 3.2_1 Download 10-year daily
df_10yr=pd.DataFrame(yf.download(tk,period='10y',interval=interval))
df_10yr
```

```python
# 3.2.2 Download 10year monthly close price, change interval as '1mo'
df_10yr=pd.DataFrame(yf.download(tk,period='10y',interval='1mo')) 
df_10yr
```

```python
# 3.2.3 Download only 'Adj Close' column
df_10yr_adj_c=pd.DataFrame(yf.download(tk,period='10y',interval='1mo')['Adj Close'])
df_10yr_adj_c
```

```python
# 3.3.1 Download from start date to today
df_from=pd.DataFrame(yf.download(tk,start=s_d,interval=interval)['Adj Close'])
df_from
```

```python
# 3.3.2 Download from start date to end date
df = pd.DataFrame(yf.download(tk,start=s_d,end=e_d,interval=interval)['Adj Close'])
df
```

4. Draw simple line chart
- Change DataFrame variable('sns.lineplot(df)') eg, 'df_10yr',  'df_ytd_adj_c'


```python
# 4.1 YTD daily Chart 
plt.figure(figsize=(10,6))
sns.lineplot(df_ytd_adj_c)
plt.ylabel(tk)
plt.title(tk + " Daily Chart YTD")

# 4.2  Daily Chart from start date to end date
plt.figure(figsize=(10,6))
sns.lineplot(df)
plt.ylabel(tk)
plt.title(tk + " Daily Chart from "+s_d +" to "+e_d)

# 4.3  10-Year monthly chart
plt.figure(figsize=(10,6))
sns.lineplot(df_10yr_adj_c)
plt.ylabel(tk)
plt.title(tk + " 10-Year Monthly Chart");
```

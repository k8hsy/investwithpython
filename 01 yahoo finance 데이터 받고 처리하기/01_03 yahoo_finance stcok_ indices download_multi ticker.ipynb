1. The process of downloading major stock indices data provided by Yahoo Finance and processing as Dataframe
    - Continue from "01_02 yahoo_finance stock_indices download historical price.ipynb"
    - Create 'Tickers' list from DataFrame column    
    - Download multi historical price data from yahoo finance
        - Input : indices ticker list
        - Output: DataFrames of multi historical data, Export Dataframe to excel file or csv file

```python
# Loading Python Packages to use 
import yfinance as yf
import pandas as pd                

# use the code from "01_02 yahoo_finance stock_indices download historical price.ipynb"

major_indices = pd.DataFrame(pd.read_html("https://finance.yahoo.com/world-indices")[0])

# Slicing columns
i = 5                                       # set column counter variable, change i to slicing columns
major_indices1 = major_indices.iloc[:, :i]  # create another dataframe 'major_indices1'
major_indices1
drop_row_index_list = [4,5,7,11,12,13,14,20,25,28,31,33,34,35]  

major_indices2 = major_indices.drop(drop_row_index_list,axis=0)       
major_indices1_2 = major_indices1.drop(drop_row_index_list,axis=0) #Datafrmae to use 
```

2. Create Ticker list to use yahoo download
    - Use ['Symbol'] column in 'major_indices1_2' DataFrame
    - :   .values.tolist()

```python
Tks=major_indices1_2['Symbol'].values.tolist()
Tks
```

```python
# Declare historical price variables[symbols('ticker'), date, interval, etc]
# yf.download(ticker, start, end, interval, etc) input varibales:'ticker', 'start=', 'end=', 'interval=' etc
tk = Tks                    # Use ticker list 'Tks' of 'Symbol' from DataFrame 'major_indices1_2' 

s_d = '2019-12-20'          # start date
e_d = '2022-12-31'          # end date, if it is up to today, just drop 'end' parameter
periods = 'ytd'             # Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
interval='1d'               # Valid interval : Daily;'1D', Weekly:'1wk',monthly:'1mo', 3month :'3mo'                        
```

3. Download multi ticker data

```python
# 3.1 Download multi index price data
df_ytd =pd.DataFrame(yf.download(tk,period='ytd',interval=interval))
df_ytd
```

```python
# 3.2 Download only 'Adj Close' column
df_ytd_adj_c=pd.DataFrame(yf.download(tk,period='ytd',interval=interval)['Adj Close'])
df_ytd_adj_c
```

```python
# 3.3 Download 10-year monthly data
df_10yr_adj_c=pd.DataFrame(yf.download(tk,period='10y',interval='1mo')['Adj Close'])
df_10yr_adj_c
```

```python
# 3.4 Download from start date to end date
df = pd.DataFrame(yf.download(tk,start=s_d,end=e_d,interval=interval)['Adj Close'])
df
```

4. Export Dataframe to excel file or csv file 
    - DataFrame.to_excel('File name.xlsx')
    - DataFrame.to_csv('file name.csv')

```python
# Export to excel file

df_ytd.to_excel('01_03 df_ytd.xlsx')
df_10yr_adj_c.to_excel('01_03 df_10yr_adj_c.xlsx')
df.to_excel('01_03 df.xlsx')

# Export to csv file (csv file contain only date type from date datatype.)

df_ytd.to_csv('01_03 df_ytd.csv')
df_10yr_adj_c.to_csv('01_03 df_10yr_adj_c.csv')
df.to_csv('01_03 df.csv')
```

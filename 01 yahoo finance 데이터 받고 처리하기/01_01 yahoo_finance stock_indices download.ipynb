1. The process of downloading major stock indices data provided by Yahoo Finance and processing as Dataframe

```python
# Loading Python Packages to use 
import yfinance as yf
import pandas as pd
```

2. Input : url for major stock indices in Yahoo Finance
   url = https://finance.yahoo.com/world-indices 

   Output : Major stcok indices dataframe tables to use

```python

# Major equity indices from Yahoo Finance 
# url = https://finance.yahoo.com/world-indices 

# From URL, read first table, and create dataframe as major_indices   
first_table=pd.read_html("https://finance.yahoo.com/world-indices")[0]

# Create the first indices dataframe 'major_indices'
major_indices = pd.DataFrame(first_table)

# The two lines of code above can be implemented in one line as shown below.
# major_indices = pd.DataFrame(pd.read_html("https://finance.yahoo.com/world-indices")[0])

# Output ; major stock indices dataframe table named 'major_indices'
major_indices
```

3. Dataframe processing
    - export dataframe table to excel for further data processing
        - Input : desired excel file name, e.g "major stock indices.xlsx"
        - Output : export excel file in the same directory with this python code if not specified directory 
        
    - slicing columns and rows in Python
        - Input : column counter variable 'i'(0 ~ last column number)
        - Output : column sliced dataframe ()
                'major_indices1' : columns slicing
                'major_indices2' : rows slicing
                'major_indices1_2 : colums & rows    
        

```python
# Slicing columns
i = 5                                       # set column counter variable, change i to slicing columns
major_indices1 = major_indices.iloc[:, :i]  # create another dataframe 'major_indices1'
major_indices1
```

```python
    # Make a list of unused indices as index number of dataframe 'major_indices'
    # Change the index number below list to slicing row
drop_row_index_list = [3,4,5,7,11,12,13,14,20,25,28,31,33,34,35]  

    # New dataframe,"major_indices2" sliced from 'drop_row_index_list' 
major_indices2=major_indices.drop(drop_row_index_list,axis=0)       

    # New dataframe,"major_indices1_2" sliced columns and rows
major_indices1_2=major_indices1.drop(drop_row_index_list,axis=0)

    # show final dataframe 
major_indices1_2
```

```python
# Export to excel
major_indices1_2.to_excel("01_01 major stock indices.xlsx")
```
